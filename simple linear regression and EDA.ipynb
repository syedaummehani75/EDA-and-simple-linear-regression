```python
#converting object into numeric
df_dummies=pd.get_dummies(df["Illness"])
df_dummies.tail()
```

```python
#creation of dummy variables
df.select_dtypes(include=[np.number]).columns.tolist()
```

```python
#Creating df of Numeric columns
df_num = df[['Age','Income']]
df_num
```

```python
#list of categorical variables
df.select_dtypes(include=['object']).columns.tolist()
```

```python
df_dummies.tail()
```

```python
#column wise, appending of columns
df_combined = pd.concat([df_num, df_dummies], axis=1)
```

```python
#modified data set
df_combined.head()
```

```python
df_combined.isnull().sum(axis = 0)
```

```python
X = df_combined.drop("Age", axis=1)
```

```python
y = df_combined['Age']
```

```python
X
```

```python
y #response variable
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7 , random_state=12)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sk= LinearRegression()
sk
```

```python
sk.fit(X_train,y_train)
```

```python
print(sk.intercept_)
```

```python
print(sk.coef_)
```

```python
coeff_df = pd.DataFrame(sk.coef_,X_test.columns,columns=['Coefficient'])
coeff_df
```

```python
sns.scatterplot(df_combined['Age'],df_combined["Income"])
```

```python
#predection
y_pred = sk.predict(X_test)
print(y_pred)
```

```python
#c0st function
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
r_squared
```

```python
from math import sqrt
rmse = sqrt(mse)
print('Mean_Squared_Error :' ,mse)
print('Root_Mean_Squared_Error :' ,rmse)
print('r_square_value :',r_squared)
```
